```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[[ 'charges',]]
```

```python
dependent
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
independent=sc.fit_transform(independent)    
```

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.linear_model import LinearRegression 
param_grid ={'fit_intercept':[True,False],'copy_X':[True,False],'n_jobs':[None],'positive':[True,False]}
grid = GridSearchCV(LinearRegression (), param_grid, refit = True, verbose = 3,n_jobs=-1) 
grid.fit(independent,dependent) 
```

```python
re=grid.cv_results_

print("TheR_score valueforbestparameter{}:".format(grid.best_params_))
table=pd.DataFrame.from_dict(re)
table

```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:")) 
children_input=float(input("Children:")) 
sex_male_input=int(input("Sex Male 0 or 1:")) 
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=grid .predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
```
