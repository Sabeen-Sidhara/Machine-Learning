```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[[ 'charges',]]
```

```python
dependent
```

```python

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc=sc.fit_transform(independent)
```

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion':['squared_error','friedman_mse','poisson','absolute_error'],
              'max_features': ['sqrt','log2'],
              'n_estimators':[50,100]} 

grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose= 3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
y_predt=grid.predict(independent)
from sklearn.metrics import r2_score
r_score=r2_score(dependent,y_predt)
r_score
```

```python
re=grid.cv_results_

print("TheR_score valueforbestparameter{}:".format(grid.best_params_))
table=pd.DataFrame.from_dict(re)
table

```
